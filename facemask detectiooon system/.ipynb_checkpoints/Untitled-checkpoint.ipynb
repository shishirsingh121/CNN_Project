```python
withmask_label=[1]*3725
withoutmask_label=[0]*3828
```

```python
labels=withmask_label+withoutmask_label
```

```python
# convert images to numpy arrays+

with_mask_path = r'C:\Users\shish\OneDrive\Desktop\python code\cnn project\facemask detectiooon system\archive\data\with_mask/'

data = []

for img_file in with_mask_files:

  image = Image.open(with_mask_path + img_file)
  image = image.resize((128,128))
  image = image.convert('RGBA')
  image = np.array(image)
  data.append(image)



without_mask_path = r'C:\Users\shish\OneDrive\Desktop\python code\cnn project\facemask detectiooon system\archive\data\without_mask/'


for img_file in without_mask_files:

  image = Image.open(without_mask_path + img_file)
  image = image.resize((128,128))
  image = image.convert('RGBA')
  image = np.array(image)
  data.append(image)
```

```python
type(data)
```

```python
data[0]
```

```python
X = np.array(data)
y = np.array(labels)
```

```python
X[0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape,y_train.shape)
```

```python
print(X_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers,Sequential
```

```python
X_train_scaled = (X_train / 255).astype('float32')
X_test_scaled = (X_test / 255).astype('float32')
```

```python
model=Sequential([
    layers.Conv2D(32,(2,2),activation='relu',input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(32,(2,2),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(32,(2,2),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation="relu"),
    layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=['accuracy']
)
```

```python
model.fit(X_train_scaled,y_train,epochs=10)
```
